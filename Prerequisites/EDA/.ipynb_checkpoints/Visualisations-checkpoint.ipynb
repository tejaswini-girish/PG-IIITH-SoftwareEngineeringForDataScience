```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Load the dataset

tips_data = sns.load_dataset('tips')
tips_data
```

```python
## Line plot

X = np.linspace(1,20,100)
Y = np.exp(X)

plt.plot(X,Y)
```

```python
## to get red stars

plt.plot(X,Y, 'r*')
```

```python
## multiple lines
Y1 = X
Y2 = np.square(X)

plt.plot(X,Y1,'r',X,Y2,'b')
plt.title('Line Plot')
```

```python
## Scatter Plot
X = tips_data['total_bill']
Y = tips_data['tip']

plt.scatter(X,Y)
plt.xlabel('total_bill')
plt.ylabel('tip')

```

```python
plt.scatter(X,Y, s = np.array(Y)**2, c = 'green', alpha=0.8)
plt.xlabel('total_bill')
plt.ylabel('tip')
```

```python
## Bar Plot
tips_data.smoker.value_counts().plot(kind='bar', rot=0)

## to display count as text
plt.text(x = 0.98 , y = tips_data.smoker.value_counts()[1]+1, s = tips_data.smoker.value_counts()[1])
plt.text(x = -0.05, y = tips_data.smoker.value_counts()[0]+2, s = tips_data.smoker.value_counts()[0])
```

```python
## Pie

plt.pie(tips_data.day.value_counts().values, autopct='%.1f%%', radius=1.2, labels=tips_data.day.value_counts().index)
plt.show()
```

```python
tips_data.day.value_counts()
```

```python
## Histogram

plt.hist(tips_data['tip'], bins=5)
plt.title("Distribution")
plt.xlabel('tip')
plt.ylabel('Frequency')
```

```python
## Box Plot

plt.boxplot(tips_data['total_bill'])
plt.text(x = 1.1, y = tips_data['total_bill'].min(), s = 'min')

plt.show()
```
