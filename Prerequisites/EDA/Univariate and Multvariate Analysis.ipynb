## Univariate Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# load the csv file of data
df_suicide = pd.read_csv('suicide_data.csv')

# display the head() of the dataset
df_suicide.head(200)
```

### Consider the variable 'gdp_per_capita' for univariate analysis

#### In our data, 'gdp_per_capita' is a numerical variable. Thus, we can use different visualization techniques (histogram, boxplot, violinplot, and so on) to study the distribution of the variable.

### Histogram - [Kindly Provide the Inferences]

```python
# Plot the histogram 
sns.histplot(data = df_suicide['gdp_per_capita ($)'])

## Inferences:
#1. The count of Lower GDP per Capita is high than the count of GDP per Capita of $120000
#2. The major chunk for GDP per capita is in the range of $0 - $60000 
```

### Distribution plot: Use distplot() to plot a kernel density estimator (KDE) along with the histogram to study the distribution of the data

```python
# Plot the distribution plot
sns.distplot(df_suicide['gdp_per_capita ($)'], kde = True)

##Inferenes
#1. The density of 3 is found for GDP per capita in the range of $0 to $20000
#2. The density of 7,8 also exists for a range of lower GDP per capita
```

### Violin plot: It is similar to a boxplot which shows the distribution of the quantitative variable

```python
# plot the violin plot
sns.violinplot(x=df_suicide['gdp_per_capita ($)'])

## Inferences
# 1. This plot cleary proves that the density of GDP per capita is maximum at the Quartile 1 but its also tapering indicating that the GDP range is small
# 2. We can also see that the median is between 0 - $20000
# 3. Density is also significant around the median value
```

### Boxplot: Boxplot can be used to visualize the presence of outliers in the data. In boxplot, the observation will be an outlier, if it is outside the specific range (1.5 times IQR above the third quartile and below the first quartile)

```python
#Plot the Box plot
sns.boxplot(x=df_suicide['gdp_per_capita ($)'])

## Infereneces
#1. IQR range is between $1000 to $25000
#2. Min is $0 and max is close to $60000
#3. All the values over the maximum is considered as outliers
```

# Multivariate Analysis

### Grouped boxplot : It is used to compare the distribution of different categories in the categorical variable. Plot a grouped boxplot to check the extreme values for the population in the different generations

```python
# Plot a Groped Box plot
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.boxplot(data = df_suicide,x='population', y='generation')

## Inferences
#1. Generation X : Extreme values close to 45M
#2. Silent : 32M
#3. G.I. Generation : 22M
#4. Boomers : 48M
```
