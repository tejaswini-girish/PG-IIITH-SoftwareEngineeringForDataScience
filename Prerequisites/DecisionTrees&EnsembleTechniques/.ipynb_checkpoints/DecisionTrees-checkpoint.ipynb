```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('credit.csv')
df.head(20)
```

```python
df.describe().transpose()
```

```python
df.corr()
```

```python
for feature in df.columns:
    if df[feature].dtype == 'object':
        df[feature] = pd.Categorical(df[feature]).codes
```

```python
df
```

```python
sns.pairplot(df, hue='purpose')
```

```python
X = df.drop('default', axis = 1)
y = df[['default']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state=1)
```

```python
train_lables = ['yes','no']
dt_model = DecisionTreeClassifier(criterion='entropy')
dt_model.fit(x_train, y_train)
```

```python
## Finding important features to build this model

pd.DataFrame(dt_model.feature_importances_, columns = ["Imp"], index = x_train.columns)
```

```python
from IPython.display import Image
from sklearn import tree
from os import system
```

```python
train_char_label = ['yes','no']
Credit_Tree_File = open('credit_tree.dot','w')
dot_data = tree.export_graphviz(dt_model, out_file=Credit_Tree_File, feature_names=list(x_train), class_names=train_char_label)

Credit_Tree_File.close()
```

```python
# system("dot -Tpng credit_tree.dot -o credit_tree.png")
# Image('credit_tree.png')

## Write the content of .dot file to http://www.webgraphviz.com/ to view the decision tree
```

```python
y_predict = dt_model.predict(x_test)
```

```python
dt_model.score(x_train, y_train)
```

```python
dt_model.score(x_test, y_test)
```

```python
metrics.confusion_matrix(y_test, y_predict)
```

1. out of 86 defaulters the model was able to identify only 34
2. out of 214 non defaulters the model was able to identify 164.
3. It has performed poorly in classifying defaulters than non defaulters

## Regularize the Decision Tree

```python
reg_dt_model = DecisionTreeClassifier(criterion='entropy', max_depth=5)
reg_dt_model.fit(x_train, y_train)
```

```python
credit_tree_regularised = open('credit_tree_regularized.dot','w')
dot_data = tree.export_graphviz(reg_dt_model, out_file=credit_tree_regularised, feature_names=list(x_train), class_names=train_char_label)

credit_tree_regularised.close()
```

```python
y_predict = reg_dt_model.predict(x_test)
```

```python
reg_dt_model.score(x_train, y_train)
```

```python
reg_dt_model.score(x_test, y_test)
```

The accuracy improved significantly after regularising the decision tree based on maximum depth
